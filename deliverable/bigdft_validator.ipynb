```python
initial_function = """
def calculate_hcn_atomization_energy():
    # Importation des modules nécessaires
    from BigDFT.Systems import System
    from BigDFT.Fragments import Fragment
    from BigDFT import Calculators as C

    # Fonction pour calculer l'énergie totale d'un système
    def calculate_total_energy(system):
        study = C.SystemCalculator(verbose=False, skip=True)
        log = study.run(sys=system, name="calc", run_dir='scratch')
        return log.energy

    # Définir la molécule HCN avec les unités angströms
    HCN = System()
    HCN["H:0"] = Fragment([{ 'sym': 'H', 'r': [0.0, 0.0, 0.0], 'units': 'angstroem' }])
    HCN["C:0"] = Fragment([{ 'sym': 'C', 'r': [0.0, 0.0, 1.064], 'units': 'angstroem' }])
    HCN["N:0"] = Fragment([{ 'sym': 'N', 'r': [0.0, 0.0, 2.229], 'units': 'angstroem' }])
    
    # Définir les atomes individuels avec les unités angströms
    H = System()
    H["H:0"] = Fragment([{ 'sym': 'H', 'r': [0.0, 0.0, 0.0], 'units': 'angstroem' }])
    
    C = System()
    C["C:0"] = Fragment([{ 'sym': 'C', 'r': [0.0, 0.0, 0.0], 'units': 'angstroem' }])
    
    N = System()
    N["N:0"] = Fragment([{ 'sym': 'N', 'r': [0.0, 0.0, 0.0], 'units': 'angstroem' }])
    
    # Calculer l'énergie totale pour chaque système
    hcn_energy = calculate_total_energy(HCN)
    h_energy = calculate_total_energy(H)
    c_energy = calculate_total_energy(C)
    n_energy = calculate_total_energy(N)
    
    # Calculer l'énergie d'atomisation
    atomization_energy = (h_energy + c_energy + n_energy) - hcn_energy
    return atomization_energy

# Exemple d'utilisation de la fonction
hcn_atomization_energy = calculate_hcn_atomization_energy()
print(f"L'énergie d'atomisation de HCN est: {hcn_atomization_energy} Ha")
"""
```

```python
message = f"""
Use the remote_run_code tool to run the following python function on 'localhost':
function_source='{initial_function}, with function_args={{}}'
"""

await chat.chat(message)
```
